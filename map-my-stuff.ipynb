```python
import yaml
from bokeh.io import output_file, show, output_notebook
from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, Range1d, PanTool, WheelZoomTool, BoxSelectTool
)

with open("config.yml", 'r') as ymlfile:
    cfg = yaml.load(ymlfile)

   
map_options = GMapOptions(lat=-33.805343, lng=151.085476, map_type="roadmap", zoom=12)

plot = GMapPlot(x_range=Range1d(), y_range=Range1d(), map_options=map_options)
plot.title.text = "My Day"
plot.api_key = cfg['maps']['key']

source = ColumnDataSource(data=dict(lat=[-33.805343], lon=[151.085476] ))
circle = Circle(x="lon", y="lat", size=10, fill_color="red", fill_alpha=0.8, line_color=None)
plot.add_glyph(source, circle)

source = ColumnDataSource(data=dict(lat=[-33.838922], lon=[151.071505] ))
circle = Circle(x="lon", y="lat", size=8, fill_color="blue", fill_alpha=0.8, line_color=None)
plot.add_glyph(source, circle)

plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())
output_notebook()

show(plot)
```
